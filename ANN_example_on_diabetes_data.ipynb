## Cross validation:

```python
X_train
```

```python
y_train.shape
```

```python
def create_model(n):
    
    model = keras.Sequential()
    model.add(keras.Input(shape=(11,)))
    model.add(keras.layers.Dense(n, activation='relu', kernel_initializer='uniform'))
    model.add(keras.layers.Dense(n, activation='relu', kernel_initializer='uniform'))
    model.add(keras.layers.Dense(1, activation='sigmoid', kernel_initializer='uniform'))
    opt = keras.optimizers.Adam(learning_rate=0.01)
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics = ['accuracy'])
    return model
```

```python
X_train.shape
```

```python
from sklearn.model_selection import KFold

list_models=[create_model(50), create_model(20)]

def test_models(model, X_train, y_train):
    
    kf = KFold(5)
    acc=[]
    for train_ids, val_ids in kf.split(X_train):
        model.fit(X_train[train_ids], y_train[train_ids], batch_size=10, epochs=5)
        acc.append(model.evaluate(X_train[val_ids], y_train[val_ids])[1])
    avg_acc = sum(acc)/len(acc)
    return avg_acc

result = []
for model in list_models:
    result.append(test_models(model, X_train, y_train))
```

```python
result
```

### Grid Search CV

```python
#https://machinelearningmastery.com/how-to-tune-algorithm-parameters-with-scikit-learn/

from sklearn.model_selection import GridSearchCV, KFold
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

def create_model():
    
    model = keras.Sequential()
    model.add(keras.Input(shape=(11,)))
    model.add(keras.layers.Dense(6, activation='relu', kernel_initializer='uniform'))
    model.add(keras.layers.Dense(6, activation='relu', kernel_initializer='uniform'))
    model.add(keras.layers.Dense(1, activation='sigmoid', kernel_initializer='uniform'))
    opt = keras.optimizers.Adam(learning_rate=0.01)
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics = ['accuracy'])
    return model

kf_cv = KFold(4, shuffle=True)

epochs = [50,100]
batch_size = [20,30]

param_grid = dict(epochs=epochs, batch_size=batch_size)

model = KerasClassifier(build_fn=create_model, verbose=0)

grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=kf_cv)
grid_result = grid.fit(X_train,y_train)
```

```python
print(f'the model with the best score ({grid_result.best_score_}) is the one with params: {grid_result.best_params_}\n')
grid_result.cv_results_
```
